# FSRCNN Evaluation

## Objective

Evaluate the trained FSRCNN model on the DIV2K validation dataset using quantitative metrics (PSNR, SSIM and MSE) and visualize the reconstructed images.

```python
from pathlib import Path
import sys

project_root = Path.cwd().parent

sys.path.append(str(project_root))
```

```python
import matplotlib.pyplot as plt
import torch

from src.models.fsrcnn import FSRCNN
from src.checkpoint import load_model

from src.dataloader import get_validation_loader

from src.metrics import (
    calculate_psnr,
    calculate_ssim
)

from src.utils.image_utils import tensor_to_image
```

```python
device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

print(device)
```

```python
validation_loader = get_validation_loader()
```

```python
model = FSRCNN()

model = load_model(
    model,
    "../results/checkpoints/fsrcnn_best.pth",
    device
)
```

```python
total_psnr = 0.0
total_ssim = 0.0
total_mse = 0.0

num_batches = 0

model.eval()

with torch.no_grad():

    for lr_batch, hr_batch in validation_loader:

        lr_batch = lr_batch.to(device)
        hr_batch = hr_batch.to(device)

        prediction = model(lr_batch)

        total_psnr += calculate_psnr(
            prediction,
            hr_batch
        )

        total_ssim += calculate_ssim(
            prediction,
            hr_batch
        )

        mse = torch.mean(
            (prediction - hr_batch) ** 2
        )

        total_mse += mse.item()

        num_batches += 1
```

```python
average_psnr = total_psnr / num_batches
average_ssim = total_ssim / num_batches
average_mse = total_mse / num_batches

print("=" * 35)
print("FSRCNN Evaluation Results")
print("=" * 35)
print(f"Average PSNR : {average_psnr:.2f} dB")
print(f"Average SSIM : {average_ssim:.4f}")
print(f"Average MSE  : {average_mse:.6f}")
print("=" * 35)
```

```python
lr_batch, hr_batch = next(iter(validation_loader))

lr_batch = lr_batch.to(device)

with torch.no_grad():

    prediction = model(lr_batch)
```

```python
lr_image = lr_batch[0].cpu()
prediction_image = prediction[0].cpu()
hr_image = hr_batch[0]
```

```python
lr_image = tensor_to_image(lr_image)
prediction_image = tensor_to_image(prediction_image)
hr_image = tensor_to_image(hr_image)
```

```python
plt.figure(figsize=(18,6))

plt.subplot(1,3,1)
plt.imshow(lr_image)
plt.title("Bicubic")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(prediction_image)
plt.title("FSRCNN")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(hr_image)
plt.title("Ground Truth")
plt.axis("off")

plt.tight_layout()

plt.show()
```
